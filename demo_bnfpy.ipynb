```python
from bnfpy import bnf
from skimage.io import imread
import matplotlib.pyplot as plt
import numpy as np
```

Read the input low-resolution image and JOR 3x resonstruction:

```python

LR=imread("demo_img//monarchLRsmall.png")
JOR3x=imread("demo_img//monarchJOR3xsmall.png")
```

Run the BN filter using the same parameters as in:
    L. R. Emmendorfer; J. C. de Jesus; T.T. Schein, 2021 (submitted)
    To Sharpen or Not? A Study on the Effect of Two Simple Sharpening Filters Applied to Magnified Images.    

```python
BNF3x=bnf(JOR3x,LR,2,0.64)
```

Show the results:

```python
HR=imread("demo_img//monarchHRsmall.png")
fig, ((ax1, ax2, ax3)) = plt.subplots(nrows=1, ncols=3, figsize=(15, 5),
                        subplot_kw={'xticks': [], 'yticks': []})
ax1.imshow(HR)
ax1.set_title('Original (HR)')
ax2.imshow(JOR3x)
ax2.set_title('JOR 3x reconstruction (no filter)')
ax3.imshow(BNF3x)
ax3.set_title('JOR 3x reconstruction + BNF')
```
